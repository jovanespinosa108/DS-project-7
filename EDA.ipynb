```python
# Experimentos de visualizacion

# Histograma
import pandas as pd 
import plotly.express as px

car_data = pd.read_csv('vehicles_us.csv')

fig = px.histogram(car_data, x='odometer')
fig.show()
```

```python
# Grafico de dispersion
fig = px.scatter(car_data, x='odometer', y='price')
fig.show()

print(car_data.info())
print(car_data.columns)
display(car_data.head())
```

```python
# Grafico de burbujas
car_data = car_data.dropna(subset=['cylinders'])

fig = px.scatter(
    car_data,
    x='odometer',
    y='price',
    size='cylinders',
    color='condition',
    hover_name='model',
    log_x=True,
    size_max=60
    )
fig.show()
```
